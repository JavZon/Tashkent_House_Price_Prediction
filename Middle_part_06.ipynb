We create our **Transformer**

```python
import sklearn
from sklearn.base import BaseEstimator, TransformerMixin
rooms_ix, size_ix= 1, 2

class CustomTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
      pass
    def fit(self, X, y=None):
      return self
    def transform(self, X):
      rooms_per_size=X[:, rooms_ix] / X[:, size_ix]
      size_per_rooms=X[:, size_ix] / X[:, rooms_ix]
      return np.c_[X, rooms_per_size, size_per_rooms]
```

Prepare **Numerical Pipeline**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

numeric_pipeline=Pipeline([
    ('custom_obj_adder', CustomTransformer()),
    ('std_scaler', StandardScaler())
])
```

```python
prepared_num=numeric_pipeline.fit_transform(X_num)
prepared_num[:3,:]
```

Finally, **full_pipeline**

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

num_cols=['rooms', 'size', 'level', 'max_levels']
string_cols=['district']

full_pipeline=ColumnTransformer([
    ('numeric', numeric_pipeline, num_cols),
    ('string_cat',OneHotEncoder(handle_unknown='ignore'), string_cols)
])
```
